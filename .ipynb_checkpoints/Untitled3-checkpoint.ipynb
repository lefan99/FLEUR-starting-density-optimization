```python
def save_json(res ,results , file_name):
    file = open(file_name , 'w')

    atom_dict = {}
    bravais_dict = {}
    orbit_dict = {}
    orbit_dict[str(res[-1])] = results
    bravais_dict[str(res[1])] = orbit_dict
    atom_dict[ str( res[2])] = bravais_dict
    atom_dict
    json.dump(atom_dict , file) 
```
