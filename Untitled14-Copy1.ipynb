```python

orbital_number = 0

max_occ = 0

for dics in valence_orbs:
    
    orbital_number += len( dics['valenceConfig'])
    
    max_occ += sum( dics['constraints'] ) * dics['atomNumber']
```

```python
def p( l , per ): #permutation function for a given list or array l with a given permutation per
    
    return [l[i] for i in per]

def p_r(l , per):  # reverse the permutationfunction to get the orgianl array : p_r(p(list , per) , per) = list
    
    temp = list(range(len(per)))
    
    for i in range(len(temp)):
        temp[per[i]] = l[i]

    return temp
```

```python
gridsize = 50

for i in range(gridsize):
    
    rest = number_valence_electrons/2
    
    atom_list = list(range(len(valence_orbs)))
    
    random.shuffle(atom_list)
    
    orbital_counter = 0
    
    remaining_orb = max_occ
    
    print(atom_list)
    
    for j in atom_list:
    
        
        sample = []
        
        permutation = list(range(len(valence_orbs[j]['constraints'])))
        
        random.shuffle(permutation)
        
        constr_p = p( valence_orbs[j]['constraints'] , permutation )
        
        for k in range(len(valence_orbs[j]['constraints'])):
            
            orbital_counter += 1
            
            remaining_orb -= constr_p[k] * valence_orbs[j]['atomNumber']
            
            occ = 0 
            
            if rest == 0:
                
                occ = 0
                
            elif orbital_counter == orbital_number: 
                
                occ = rest/valence_orbs[j]['atomNumber']
                
                
            elif constr_p[k] * valence_orbs[j]['atomNumber']  < rest:
                
                if remaining_orb < rest:
                    
                    occ = np.random.uniform((rest - remaining_orb)/valence_orbs[j]['atomNumber'] , constr_p[k] ) 
                    
                else:
                
                    occ = np.random.uniform( 0 , constr_p[k] ) 
                
            else: 
                
                if remaining_orb < rest:
                    
                    occ = np.random.uniform((rest - remaining_orb)/valence_orbs[j]['atomNumber'] , rest/valence_orbs[j]['atomNumber'] ) 
                    
                else:
                
                    occ = np.random.uniform( 0 , rest / valence_orbs[j]['atomNumber'] ) 
            
            sample.append(occ)
            
            rest -= occ * valence_orbs[j]['atomNumber']
            

            
            
            
            #print(rest , constr_p[k] , occ , orbital_counter , remaining_orb) 
            
        valence_orbs[j]['samplesGrid'].append(p_r(sample, permutation))
        
            
    
```
