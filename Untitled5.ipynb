```python
def save_json(res ,results , file_name):
    file = open(file_name , 'w')

    atom_dict = {}
    bravais_dict = {}
    orbit_dict = {}
    orbit_dict[str(res[-1])] = results
    bravais_dict[str(res[1])] = orbit_dict
    atom_dict[ str( res[2])] = bravais_dict
    atom_dict
    json.dump(atom_dict , file) 
```

```python
def compare_json(json_file, destination):
    
    import json

    file = open(json_file , 'r' ) 

    dic = json.load(file)
    
    empty_dict = {}

    file2 = open('standard_calc_itmax1.json', 'r')

    stand = json.load(file2)

    for number in dic:
    
        for bravais in dic[number]:
        
            for orbital in dic[number][bravais]:
            
                for occupation in dic[number][bravais][orbital]:
                    
                    if type(dic[number][bravais][orbital][occupation]) == float:
                        empty_dict[occupation] = dic[number][bravais][orbital][occupation] - stand[number]
                    

               #     print(dic[number][bravais][orbital][occupation] , stand[number])
                    else:
                        empty_dict[occupation] = max(dic[number][bravais][orbital][occupation]) - stand[number]
                    
    file3 = open(destination , 'w')
    
    json.dump(empty_dict , file3)
```

```python
table= [0] * 112
import json

results_json = [ 'results_2861.json' ,'results_2862.json' ,'results_2863.json' ,'results_2864.json' ,'results_2865.json' ,'results_2866.json' ,'results_2870.json' , 'results_2871.json' ]

for file in results_json:
    
    file1 = open(file , 'r')
    
    data = json.load(file1)
    
    file2 = open('standard_calc_itmax1.json', 'r')
    
    data2 = json.load(file2)
    
    for atom in data:
        
        number = eval(atom)
        print(number)
        
        for bravais in data[atom]:
            
            for valence in data[atom][bravais]:
                
                first = True
                val = valence
                for occupation in data[atom][bravais][valence]:
                    
                    if first:
                        
                        distance = data[atom][bravais][valence][occupation]
                        
                        occ = occupation
                        
                        first = False
                        
                    else:
                        
                        if distance > data[atom][bravais][valence][occupation]:
                            
                            distance = data[atom][bravais][valence][occupation]

                            occ = occupation
                            
                            
    table[number] = [{'standard occupation and distance' : [starting_occupation_stand[number] ,  data2[str(number)]], 'minimum occupation' : [occupation , distance , val]  } ]                   
                            
    
```

```python
import pandas as pd

data_dict = {}

file = open('constraints.json' , 'r') 

constr = json.load(file)

for i in range(len(table)):
        
    if table[i] != 0:
        
        stand_state = []
        
        stand_occupation = []
        
        counter = 0
        
        for dict1 in table[i][0]['standard occupation and distance'][0]:
            
            stand_state.append(dict1['state'])
            
            
        for valence in  eval(table[i][0]['minimum occupation'][2]):
            
            if valence in stand_state:
            
                if table[i][0]['standard occupation and distance'][0][counter]['spinUp'] == table[i][0]['standard occupation and distance'][0][counter]['spinDown']:
                
                    stand_occupation.append( table[i][0]['standard occupation and distance'][0][counter]['spinUp'])
                
                else:
            
                    stand_occupation.append( (table[i][0]['standard occupation and distance'][0][counter]['spinUp'] , table[i][0]['standard occupation and distance'][0][counter]['spinDown']))
        
            
                counter += 1
            
            else:
                stand_occupation.append(constr[valence]/2)
        
        
        data_dict[str(i)] = [ table[i][0]['minimum occupation'][2] , table[i][0]['minimum occupation'][0] , table[i][0]['minimum occupation'][1] , stand_occupation , table[i][0]['standard occupation and distance'][1]]
    
        
```

```python
df = pd.DataFrame(data_dict , index = ['valence' , 'occupation' , 'distance'  , 'standard occupation' ,  'stand distance'])
```
